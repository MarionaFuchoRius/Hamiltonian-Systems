```python
import numpy as np
import matplotlib.pyplot as plt

# Definition of Parameters
gamma = 1.0
tau   = 1.0
dt    = 0.01
t_end = 10.0

# Number of steps from t=0 to t=t_end
n_steps = int(t_end/dt)

# How many steps correspond to the delay 
delay_steps = int(tau/dt)

# We create vectors to store time and y
t = np.arange(-tau, t_end + dt, dt)
y = np.zeros_like(t)

#    The problem states y(t<=0) = 1, so we set these initial values.
for i in range(delay_steps):
    y[i] = 1.0

# We implement Forward Euler method
#
#    y[n+1] = y[n] + dt * (-gamma * y[n - delay_steps])
#
for i in range(delay_steps, len(t) - 1):
    y[i+1] = y[i] - dt * gamma * y[i - delay_steps]

# Plot the result
plt.plot(t, y, label='y(t)')
plt.axvline(x=0, color='k', linestyle='--', alpha=0.7)
plt.axhline(y=0, color='k', linestyle='--', alpha=0.7)
plt.title("Solution of the given DDE via Forward Euler")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
```
